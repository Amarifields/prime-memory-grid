```python
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
import math

class PrimeMemoryGrid:
    def __init__(self, size=8):
        self.size = size
        self.grid = np.zeros((size, size), dtype=int)
        self.primes = self._generate_primes(1000)
        self.symbol_map = self._create_symbol_map()
        self.connections = defaultdict(list)

    def _generate_primes(self, limit):
        """Generate prime numbers up to limit using Sieve of Eratosthenes"""
        sieve = [True] * (limit + 1)
        sieve[0] = sieve[1] = False
        for i in range(2, int(limit**0.5) + 1):
            if sieve[i]:
                for j in range(i*i, limit + 1, i):
                    sieve[j] = False
        return [i for i in range(2, limit + 1) if sieve[i]]

    def _create_symbol_map(self):
        """Map common symbols to prime numbers"""
        symbols = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789 .,!?-'
        return {symbols[i]: self.primes[i] for i in range(min(len(symbols), len(self.primes)))}

    def encode_text(self, text):
        """Encode text into the prime grid using mathematical positioning"""
        text = text.upper()
        positions = []

        for i, char in enumerate(text):
            if char in self.symbol_map:
                prime = self.symbol_map[char]

                x = (prime * (i + 1)) % self.size
                y = (prime + i * 7) % self.size

                self.grid[x, y] = prime
                positions.append((x, y, char, prime))

        return positions

    def detect_harmony(self):
        """Detect mathematical harmony patterns in the grid"""
        harmony_score = 0
        patterns = []


        for i in range(self.size):
            row_product = np.prod([x for x in self.grid[i] if x > 0])
            col_product = np.prod([x for x in self.grid[:, i] if x > 0])

            if row_product > 1:
                factors = self._prime_factors(row_product)
                if len(set(factors)) < len(factors):
                    harmony_score += 10
                    patterns.append(f"Row {i}: Repeated prime factors {set(factors)}")


        for i in range(self.size):
            row = [x for x in self.grid[i] if x > 0]
            if len(row) > 1 and row == row[::-1]:
                harmony_score += 20
                patterns.append(f"Row {i}: Palindromic sequence {row}")


        for i in range(self.size-1):
            for j in range(self.size-1):
                if self.grid[i,j] > 0 and self.grid[i+1,j+1] > 0:
                    ratio = self.grid[i+1,j+1] / self.grid[i,j]
                    if 1.6 < ratio < 1.62:
                        harmony_score += 15
                        patterns.append(f"Golden ratio at ({i},{j}): {ratio:.3f}")

        return harmony_score, patterns

    def detect_symmetry(self):
        """Detect symmetrical patterns in the grid"""
        symmetries = []


        if np.array_equal(self.grid, np.flipud(self.grid)):
            symmetries.append("Perfect horizontal symmetry")


        if np.array_equal(self.grid, np.fliplr(self.grid)):
            symmetries.append("Perfect vertical symmetry")


        if np.array_equal(self.grid, self.grid.T):
            symmetries.append("Perfect diagonal symmetry")


        for rotation in [1, 2, 3]:
            rotated = np.rot90(self.grid, rotation)
            if np.array_equal(self.grid, rotated):
                symmetries.append(f"Rotational symmetry ({rotation * 90}¬∞)")

        return symmetries

    def detect_anomalies(self):
        """Detect mathematical anomalies and interesting patterns"""
        anomalies = []


        for i in range(self.size):
            for j in range(self.size):
                if self.grid[i,j] > 0:
                    neighbors = self._get_neighbors(i, j)
                    if all(self.grid[ni, nj] == 0 for ni, nj in neighbors):
                        anomalies.append(f"Isolated prime {self.grid[i,j]} at ({i},{j})")


        clusters = self._find_clusters()
        for cluster in clusters:
            if len(cluster) >= 3:
                anomalies.append(f"Prime cluster of size {len(cluster)}: {cluster}")

        return anomalies

    def _get_neighbors(self, x, y):
        """Get valid neighbor coordinates"""
        neighbors = []
        for dx in [-1, 0, 1]:
            for dy in [-1, 0, 1]:
                if dx == 0 and dy == 0:
                    continue
                nx, ny = x + dx, y + dy
                if 0 <= nx < self.size and 0 <= ny < self.size:
                    neighbors.append((nx, ny))
        return neighbors

    def _find_clusters(self):
        """Find connected components of non-zero values"""
        visited = np.zeros((self.size, self.size), dtype=bool)
        clusters = []

        def dfs(x, y, cluster):
            if visited[x, y] or self.grid[x, y] == 0:
                return
            visited[x, y] = True
            cluster.append((x, y, self.grid[x, y]))

            for nx, ny in self._get_neighbors(x, y):
                if not visited[nx, ny] and self.grid[nx, ny] > 0:
                    dfs(nx, ny, cluster)

        for i in range(self.size):
            for j in range(self.size):
                if not visited[i, j] and self.grid[i, j] > 0:
                    cluster = []
                    dfs(i, j, cluster)
                    if cluster:
                        clusters.append(cluster)

        return clusters

    def _prime_factors(self, n):
        """Get prime factors of a number"""
        factors = []
        d = 2
        while d * d <= n:
            while n % d == 0:
                factors.append(d)
                n //= d
            d += 1
        if n > 1:
            factors.append(n)
        return factors

    def evolve_connections(self):
        """Evolve connections based on mathematical resonance"""
        self.connections.clear()

        for i in range(self.size):
            for j in range(self.size):
                if self.grid[i, j] > 0:
                    prime1 = self.grid[i, j]


                    for x in range(self.size):
                        for y in range(self.size):
                            if (x, y) != (i, j) and self.grid[x, y] > 0:
                                prime2 = self.grid[x, y]


                                strength = self._calculate_connection_strength(prime1, prime2, i, j, x, y)
                                if strength > 0.5:
                                    self.connections[(i, j)].append(((x, y), strength))

    def _calculate_connection_strength(self, p1, p2, x1, y1, x2, y2):
        """Calculate connection strength between two primes"""
        strength = 0

        if math.gcd(p1, p2) > 1:
            strength += 0.3


        distance = math.sqrt((x2-x1)**2 + (y2-y1)**2)
        strength += max(0, 0.5 - distance/10)


        if abs(p1 - p2) in self.primes[:20]:
            strength += 0.4

        if (p1 + p2) in self.primes:
            strength += 0.3

        return min(strength, 1.0)

    def suggest_mutations(self):
        """Suggest mutations to increase mathematical harmony"""
        suggestions = []
        current_harmony, _ = self.detect_harmony()


        non_zero_positions = [(i, j) for i in range(self.size) for j in range(self.size) if self.grid[i, j] > 0]

        for i, (x1, y1) in enumerate(non_zero_positions[:5]):
            for x2, y2 in non_zero_positions[i+1:6]:

                original_grid = self.grid.copy()
                self.grid[x1, y1], self.grid[x2, y2] = self.grid[x2, y2], self.grid[x1, y1]

                new_harmony, _ = self.detect_harmony()
                if new_harmony > current_harmony:
                    suggestions.append(f"Swap positions ({x1},{y1}) and ({x2},{y2}) for +{new_harmony-current_harmony} harmony")


                self.grid = original_grid

        return suggestions[:3]

    def visualize(self):
        """Create visual representation of the grid"""
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

        im = ax1.imshow(self.grid, cmap='viridis', interpolation='nearest')
        ax1.set_title('Prime Memory Grid')


        for i in range(self.size):
            for j in range(self.size):
                if self.grid[i, j] > 0:
                    ax1.text(j, i, str(self.grid[i, j]), ha='center', va='center',
                            color='white', fontweight='bold', fontsize=8)

        plt.colorbar(im, ax=ax1)


        ax2.imshow(np.zeros((self.size, self.size)), cmap='gray', alpha=0.3)
        ax2.set_title('Connection Network')


        for (x1, y1), connections in self.connections.items():
            for (x2, y2), strength in connections:
                ax2.plot([y1, y2], [x1, x2], 'r-', alpha=strength, linewidth=strength*3)
                ax2.plot(y1, x1, 'bo', markersize=8)
                ax2.plot(y2, x2, 'ro', markersize=6)

        plt.tight_layout()
        plt.show()

    def analyze(self, text):
        """Complete analysis of encoded text"""
        print(f"üß† PRIME MEMORY GRID ANALYSIS")
        print(f"{'='*50}")
        print(f"Input: '{text}'")
        print(f"Grid Size: {self.size}x{self.size}")


        positions = self.encode_text(text)
        print(f"\nüìç ENCODING:")
        for x, y, char, prime in positions:
            print(f"  '{char}' ‚Üí Prime {prime} at ({x},{y})")


        harmony_score, harmony_patterns = self.detect_harmony()
        symmetries = self.detect_symmetry()
        anomalies = self.detect_anomalies()

        print(f"\nüéµ HARMONY ANALYSIS (Score: {harmony_score}):")
        for pattern in harmony_patterns:
            print(f"  ‚Ä¢ {pattern}")

        print(f"\nüîÑ SYMMETRY PATTERNS:")
        for sym in symmetries:
            print(f"  ‚Ä¢ {sym}")
        if not symmetries:
            print("  ‚Ä¢ No perfect symmetries detected")

        print(f"\n‚ö° ANOMALIES:")
        for anomaly in anomalies:
            print(f"  ‚Ä¢ {anomaly}")


        self.evolve_connections()
        total_connections = sum(len(conns) for conns in self.connections.values())
        print(f"\nüîó CONNECTIONS: {total_connections} mathematical links formed")


        suggestions = self.suggest_mutations()
        print(f"\nüí° OPTIMIZATION SUGGESTIONS:")
        for suggestion in suggestions:
            print(f"  ‚Ä¢ {suggestion}")
        if not suggestions:
            print("  ‚Ä¢ Grid is optimally arranged")


        self.visualize()

        return {
            'harmony_score': harmony_score,
            'patterns': harmony_patterns,
            'symmetries': symmetries,
            'anomalies': anomalies,
            'connections': total_connections,
            'suggestions': suggestions
        }


print("üöÄ PRIME MEMORY GRID ALGORITHM DEMO")
print("="*60)


pmg = PrimeMemoryGrid(size=6)


test_cases = [
    "HELLO WORLD",
    "MATHEMATICS",
    "PRIME GRID"
]

for i, test_text in enumerate(test_cases, 1):
    print(f"\nüî¨ TEST CASE {i}")
    print("-" * 40)


    pmg.grid = np.zeros((pmg.size, pmg.size), dtype=int)


    results = pmg.analyze(test_text)

    print(f"\nüìä SUMMARY:")
    print(f"  Harmony Score: {results['harmony_score']}")
    print(f"  Patterns Found: {len(results['patterns'])}")
    print(f"  Symmetries: {len(results['symmetries'])}")
    print(f"  Anomalies: {len(results['anomalies'])}")
    print(f"  Connections: {results['connections']}")

print(f"\n‚ú® Prime Memory Grid Algorithm Complete!")
print(f"üìà Novel intelligence through pure mathematical operations")
print(f"üî¢ Based on number theory and symbolic compression")
```
